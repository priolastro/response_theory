```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio
```

```python
def sin_wave(f,a,duration,sample_rate):
    t=np.arange(0,duration,1/sample_rate)
    return a*np.sin(2*np.pi*f*t), t
```

```python
s, t=sin_wave(f=261.63,
             a=1,
             duration=0.25,
             sample_rate=10e3)
```

```python
plt.plot(t[:80],s[:80])
```

```python
fs=10e3
Audio(s,rate=fs)
```

```python
f=np.array([261.6,329.6,392.0])
a=np.array([1.5,0.5,1])
s,t=sin_wave(f=f.reshape(3,1),
            a=a.reshape(3,1),
            duration=0.25,
            sample_rate=10e3)
```

```python
plt.plot(t[:80],s[0][:80])
plt.plot(t[:80],s[1][:80])
plt.plot(t[:80],s[2][:80])
```
